use the plotly library to make visualizations

```python
import plotly
```

using the numpy library

```python
import numpy as np
```

```python
import pandas as pd
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)
```

create a new column

```python
df_2014_2015_concat["marker_color"] = np.where(
    df_2014_2015_concat.order_year == "2014", # if order_year == 2014, it will be colored green
    "green",
    "red" # if order_year == 2015, it will be colored red
)

df_2014_2015_concat
```

```python
df_2014_2015_concat.sort_values("order_year", ascending=True)
```

```python
df_2014_2015_concat
```

make a scatter plot

```python
data = go.Scatter(
    x=df_2014_2015_concat.sales,
    y=df_2014_2015_concat.profit,
    mode="markers",
    marker={
        "color": df_2014_2015_concat.marker_color,
    },
    name="Sales VS Profit 2014-2015"
)

# add layout variables
layout = {
    "title": {
        "text" : "Sales VS Profit 2014-2015",
        "x": 0.5, # the value x has values from 0 to 1. if using 0.5 it is in the middle
    },
    "xaxis": { # add labels on the x axis
        "title": "Sales",
    },
    "yaxis": { # add labels on the y axis
        "title": "Provit",
    }
}

# we create an object figure

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
dot1 = go.Scatter(
    x=df_2014_2015_concat[df_2014_2015_concat.order_year == "2014"].sales,
    y=df_2014_2015_concat[df_2014_2015_concat.order_year == "2014"].profit,
    mode="markers",
    marker={ # add marker parameters
        "size": 6,
    },
    name="2014"
)

dot2 = go.Scatter(
    x=df_2014_2015_concat[df_2014_2015_concat.order_year == "2015"].sales,
    y=df_2014_2015_concat[df_2014_2015_concat.order_year == "2015"].profit,
    mode="markers",
    marker={ # add marker parameters
        "size": 12,
    },
    name="2015"
)

data = [dot1,dot2]

# add layout variables
layout = {
    "title": {
        "text" : "Sales VS Profit 2014-2015",
        "x": 0.5, # x values have values from 0 to 1. if you use 0.5, it is in the middle
    },
    "xaxis": { # add labels on the x axis
        "title": "Sales",
    },
    "yaxis": { # add labels on the y axis
        "title": "Provit",
    }
}

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
